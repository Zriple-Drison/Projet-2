## Choisissons des indicateurs pertinents 

Pour répondre à la problématique,on a besoin de savoir le nombre de personnes qui ont accès à internet dans le pays, le pib par habitant, le nombre d'élèves qui vont aller au lycée et ou dans le supérieur.

### Internet

```python
def find_indicator(theme_indicateur,theme_indicateur2):
    
    df_series = series_clean[(series_clean['Topic'].str.contains(theme_indicateur)) |
                             (series_clean['Topic'].str.contains(theme_indicateur2)) |
                             (series_clean['Indicator Name'].str.contains(theme_indicateur)) |
                             (series_clean['Indicator Name'].str.contains(theme_indicateur2)) |
                             (series_clean['Short definition'].str.contains(theme_indicateur)) |
                             (series_clean['Short definition'].str.contains(theme_indicateur2)) |
                             (series_clean['Long definition'].str.contains(theme_indicateur)) |
                             (series_clean['Long definition'].str.contains(theme_indicateur2))]
    
    return df_series
```

### Resultats des indicateurs par pays

Pour internet, cherchons les données des taux d'accès à internet les plus récents possibles.

```python
data_it=data_clean2[data_clean2['Indicator Code']== 'IT.NET.USER.P2']
```

```python
data_it['last_year'] = None

# 3. Parcourir les lignes et mettre à jour la colonne 'last_year'
for index, row in data_it.iterrows():
    last_valid_year = None
    for year in range(2016, 2000, -1):  # Parcours des années de 2016 à 2001
        if not pd.isnull(row[str(year)]):
            last_valid_year = year
            break
    data_it.at[index, 'last_year'] = last_valid_year
```

```python
sns.countplot(x=data_it.last_year.sort_values(), color='lightgreen', edgecolor='black')
plt.title('Nombre de pays avec la dernière année où il y a une donnée pour accès à internet');
```

Pour la majorité des pays la dernière donnée date de 2016, on prend donc 2016 comme last year.

```python
meanintdata = data_it['2016'].mean()
data_it_filtered = data_it[data_it['2016'] > meanintdata] #on enlève tout les pays avec un taux d'accès internet inférieur à la moyenne de 2016
```

```python
data_it_filtered2 = data_it_filtered.loc[:,("Country Name","Country Code", "Indicator Name", "Indicator Code", "2016", "last_year")]
```

```python
data_it_filtered2.rename(columns={'2016':'last data'}, inplace=True)
```

```python
data_it_filtered2_sorted = data_it_filtered2.sort_values(by='last data', ascending=False)
```

```python
ax = data_it_filtered2_sorted.plot(kind = 'bar',x='Country Name', y='last data',
                               xlabel = 'Pays', ylabel = "% internet acces",
                               figsize = [20,5], fontsize=11)
ax.get_legend().remove()
plt.title('Percentage of population with internet acces')
```
